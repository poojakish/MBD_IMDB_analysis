```python
import pandas as pd

file_path = 'DirectorPopularityMovies.csv'
df_actor = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
```

```python
df_actor
```

```python
df_Sorted=df_actor.sort_values(by='average_popularity', ascending=False)
df_Filtered= df_Sorted[df_Sorted['birthYear']!="\\N"]
```

```python
df_Filtered['birthYear'] = pd.to_numeric(df_Filtered['birthYear'], errors='coerce')
df_Filtered= df_Filtered[df_Filtered['birthYear']>1950]
df_Filtered = df_Filtered[df_Filtered['average_popularity']>5.5]
df_Filtered
```

```python
df_Filtered['aggregated_genres'] = df_Filtered['aggregated_genres'].str.replace(', ',',')
```

```python
df_Filtered['aggregated_genres'] = df_Filtered['aggregated_genres'].str.split(',')
df_Filtered
```

```python
df_filtered_genres= df_Filtered[df_Filtered['aggregated_genres'].apply(lambda x: len(x) > 20)]
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_Filtered['primaryName'], df_Filtered['average_popularity'], s=df_Filtered['average_popularity']*20, c=df_Filtered['average_popularity'], cmap='viridis')

# Add labels and title
plt.xlabel('Actor')
plt.ylabel('Average Popularity')
plt.title('Bubble Chart of Actors and Average Popularity')

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Average Popularity', rotation=270, labelpad=15)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df_filtered_genres= df_filtered_genres.drop(columns=['birthYear','deathYear','primaryProfession','nconst'])
# df_filteres_genres = df_Filtered.drop(columns=['birthYear','deathYear','primaryProfession','nconst'])
```

```python

df_filtered_genres1=df_Filtered[:20]
df_exploded = df_filtered_genres1.explode('aggregated_genres')

df_filtered = df_exploded[~df_exploded['aggregated_genres'].str.strip().isin(['\\N', '\\\\N','Short','Sport','News','Talk-Show'])]

# Group by actor and genre, count occurrences
actor_genre_distribution = df_filtered.groupby(['primaryName', 'aggregated_genres']).size().reset_index(name='count')

# Calculate the total count of movies for each actor
actor_total_movies = actor_genre_distribution.groupby('primaryName')['count'].sum().reset_index(name='TotalMovies')

# Calculate the top 10 actors based on total movie count
top_10_actors = actor_total_movies.nlargest(10, 'TotalMovies')

# Merge the top 10 actors with the genre distribution data
top_10_genre_distribution = pd.merge(top_10_actors, actor_genre_distribution, on='primaryName')

# Pivot the data for plotting
pivot_df = top_10_genre_distribution.pivot(index='primaryName', columns='aggregated_genres', values='count').fillna(0)

# Plot the bar chart
pivot_df.plot(kind='bar', stacked=True)

plt.xlabel('Directors')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution for Top 10 Directors')
plt.legend(title='Genre', bbox_to_anchor=(1, 1))
plt.show()
```
