```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'RegionPopularity.csv'
df_PR = pd.read_csv(file_path)
```

```python
df_PR
```

```python
plt.scatter(df_PR['unique_regions_count'], df_PR['avg_popularity_index'])
plt.title('Popularity vs Region Counts')
plt.xlabel('Region Counts')
plt.ylabel('Popularity')
plt.show()
```

```python
plt.bar(df_PR['unique_regions_count'], df_PR['avg_popularity_index'], color='skyblue')
plt.title('Popularity by Region')
plt.xlabel('Region')
plt.ylabel('Popularity')
plt.show()
```

```python
df_filtered=df_PR[df_PR['unique_regions_count']<80]
plt.bar(df_filtered['unique_regions_count'], df_filtered['avg_popularity_index'], color='skyblue')
plt.title('Popularity by Region')
plt.xlabel('Region')
plt.ylabel('Popularity')
plt.show()
```

```python
bins = list(range(0, 101, 10))
df_PR['Range_unique_regions_count'] = pd.cut(df_PR['unique_regions_count'], bins=bins, labels=[f"{i}-{i+9}" for i in range(0, 100, 10)])

# Calculate average values for each range
avg_values = df_PR.groupby('Range_unique_regions_count')['avg_popularity_index'].mean().reset_index()
```

```python
avg_values
```

```python
plt.bar(avg_values['Range_unique_regions_count'], avg_values['avg_popularity_index'], color='skyblue')
plt.title('Popularity by Region release count')
plt.xlabel('Region Count')
plt.ylabel('Popularity')
plt.show()
```
